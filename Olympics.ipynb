```python
def fetch_medal_tally(years,country):
    medal_df =df.drop_duplicates(subset=['Team','NOC','Games','Year','City','Sport','Event','Medal'])
    flag=0
    if years == 'overall' and country == 'overall':
      temp_df = medal_df
    if years == 'overall' and country != 'overall':
      flag=1
      temp_df = medal_df[medal_df['region']==country]
    if years != 'overall' and country == 'overall':
      temp_df = medal_df[medal_df['Year']==years]
    if years != 'overall' and country != 'overall':
      temp_df = medal_df[(medal_df['Year']== 2012) & (medal_df['region']=="USA")]
    if flag==1:
        x=temp_df.groupby('Year').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
    else:
         x=temp_df.groupby('region').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()


    x['total']=x['Gold']+x['Silver']+x['Bronze']
    print(x)
```

```python
fetch_medal_tally('overall','USA')
```
